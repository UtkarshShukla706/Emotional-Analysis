```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# so here i am using the emotion sentiment analysis file for the test with name of the train.txt

# hence the data is given and later it is surrounded by the ; in the text file

df=pd.read_csv('C:/Users/Acer/Emotional Analysis/train.txt',sep=';',header=None,names=['text','emotion'])
```

```python
df.head(10)
```

```python
# the data is going to contain the text and the emotion in the data
#in the two seperate number of the columns 

#now we have to check that it contains any of the null values 

df.isnull().sum()
```

```python
#so there is no null values in the data set

df['emotion'].unique()
```

```python
unique_emotions=df['emotion'].unique()

emotion_numbers={}
i=0
for emo in unique_emotions:
    emotion_numbers[emo]=i
    i+=1
df['emotion']=df['emotion'].map(emotion_numbers)
```

```python
df.dtypes
```

```python
#so in this we have converted our output variable into the numbers

df['text']=df['text'].apply(lambda x : x.lower())

#convert all the letters into the small letters 
```

```python
#Removing the punctuation in the text dataset 

import string

def remove_punc(txt):
    return txt.translate(str.maketrans('','',string.punctuation))
```

```python
df['text']=df['text'].apply(remove_punc)
df.loc[1]['text']
```

```python
#removing the number in the text data sets 

def remove_numbers(txt):
    new=""
    for i in txt:
        if not i.isdigit():
            new=new+i
    return new

df['text']=df['text'].apply(remove_numbers)
```

```python
#this is the function for removing any kind of the url using the regex

import re

def remove_urls(txt):
    pattern = r'https?://\S+|www\.\S+|\b[a-zA-Z0-9.-]+\.(com|org|net|edu|gov|io|co|in|uk|ai|me|info|biz)(/\S*)?\b'
    txt = re.sub(pattern, '', txt)
    return " ".join(txt.split())  # Remove extra spaces

df['text'] = df['text'].apply(remove_urls)
```

```python
# for removing the html tags in our text dataset

import re

def remove_html_tags(txt):
    pattern = r'<.*?>'
    txt = re.sub(pattern, '', txt)
    return txt.strip()

df['text'] = df['text'].apply(remove_html_tags)
```

```python
#for removing the emojis from the text dataset

import re

df["text"] = df["text"].apply(
    lambda x: re.sub(
        r'[\U0001F600-\U0001F64F'
        r'\U0001F300-\U0001F5FF'
        r'\U0001F680-\U0001F6FF'
        r'\U0001F1E0-\U0001F1FF'
        r'\U00002700-\U000027BF'
        r'\U000024C2-\U0001F251]+',
        '',
        x
    )
)
```

```python
#in the machine learning context we have to remove the Stopwords like is , the ,was,etc

import nltk

```

```python
from nltk.corpus import stopwords

from nltk.tokenize import word_tokenize
```

```python
#now we are going to download all the punkt and the stopwords in this

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
#now after downloading all the stopwords

stop_words=set(stopwords.words('english'))
len(stop_words)
```

```python
#now we get all the stopwords now i have to remove all the stopwords

df.loc[1]['text']
```

```python
#so function for removing the stopwords

def remove(txt):
    words=word_tokenize(txt)
    cleaned=[]
    for i in words:
        if not i in stop_words:
            cleaned.append(i)
    return ' '.join(cleaned)
```

```python
df['text']=df['text'].apply(remove)
```

```python
df.loc[1]['text']

#so here we can see that all the bug words get removed all the data of the nlp is clean
```

```python
df.head(20)
```

```python
from sklearn.model_selection import train_test_split
X=df['text']
y=df['emotion']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
#now for the evaluation of the text data we are going to use the multinomial naive bayes model

from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer

bag_vectorizer=CountVectorizer()

X_train_bow=bag_vectorizer.fit_transform(X_train)
X_test_bow=bag_vectorizer.transform(X_test)

bow_nb=MultinomialNB()

bow_nb.fit(X_train_bow,y_train)
```

```python
y_pred_nb=bow_nb.predict(X_test_bow)
accuracy_score(y_test,y_pred_nb)
```

```python
#now by using the Logistic regression 

from sklearn.linear_model import LogisticRegression

log_vectorizer=LogisticRegression(max_iter=1000)

log_vectorizer.fit(X_train_bow,y_train)
```

```python
y_pred_log=log_vectorizer.predict(X_test_bow)
accuracy_score(y_test,y_pred_log)

#so here the accuracy using the bow and the logistic regression is 88.87
```

```python
#now by using the svc for the classification problem 

from sklearn.svm import LinearSVC

svm_vectorizer=LinearSVC(random_state=42)
svm_vectorizer.fit(X_train_bow,y_train)
```

```python
y_pred_svc=svm_vectorizer.predict(X_test_bow)
accuracy_score(y_test,y_pred_svc)

#so here the accuracy score is 88.96 using the svc
```

```python
#now i am going to apply each of the technique and create the better model using the tf idf method

tf_vectorizer=TfidfVectorizer()

X_train_tf=tf_vectorizer.fit_transform(X_train)
X_test_tf=tf_vectorizer.transform(X_test)
```

```python
#so lets test with the naive bayes 

tf_nb=MultinomialNB()
tf_nb.fit(X_train_tf,y_train)
```

```python
y_pred_tf_nb=tf_nb.predict(X_test_tf)
accuracy_score(y_test,y_pred_tf_nb)

#so it is 66.09
```

```python
#now for the logistic regression 

log_tf=LogisticRegression(max_iter=1000)
log_tf.fit(X_train_tf,y_train)

y_pred_tf_log=log_tf.predict(X_test_tf)
accuracy_score(y_test,y_pred_tf_log)

#so this is around 86.15

```

```python
svm_tf=LinearSVC(random_state=42)
svm_tf.fit(X_train_tf,y_train)

y_pred_tf_svm=svm_tf.predict(X_test_tf)
accuracy_score(y_test,y_pred_tf_svm)

#so it is 89.18
```

```python
#the best model we get is LinearSVC + TfIDFVectorizer
```
